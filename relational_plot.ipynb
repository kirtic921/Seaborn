- better aesthetics
- more graphs included

### Seaborn Stuff:
##### Types of functions
- Figure Level
- Axis Level

##### Main Classification
- Relational Plot  -> scatterplot, lineplot
- Distribution Plot -> histogram, kdeplot, rugplot
- Categorical Plot  -> barplot, countplot, boxplot, violinplot, swarmplot 
- Regression Plot   -> 
- Matrix Plot        -> heatmap, clustermap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

#### Relational Plot  -> scatterplot, lineplot
- to study statistical relation between two or more variables

```python
# scatter plot -> axis level

tips = sns.load_dataset('tips')
sns.scatterplot(data= tips, x='total_bill', y='tip')
```

```python
tips
```

```python
# relplot -> fig level

sns.relplot(data= tips, x='total_bill', y='tip')
```

```python
# scatter using relplot
sns.relplot(data= tips, x='total_bill', y='tip', kind='scatter')
```

```python
# hue, style, size

sns.scatterplot(data= tips, x='total_bill', y='tip', hue='sex', style='time', size = 'size')
```

```python
sns.relplot(data= tips, x='total_bill', y='tip', hue='sex', style='time', size = 'size')
```

```python
# line plot

gap = px.data.gapminder()
gap
```

```python
temp_df = gap[gap['country']=='India']
temp_df
```

```python
# axes level
sns.lineplot(data=temp_df, x='year', y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style

temp_df = gap[gap['country'].isin(['India', 'Pakistan', 'China'])]
temp_df
```

```python
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line', hue='country', style = 'continent')
```

```python
sns.lineplot(data=temp_df, x='year', y='lifeExp', hue='country', style='continent')
```

```python
# facet plot -> works with fig level functions only
# not works with lineplot and scatterplotS

sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex')
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='smoker',row='sex')
```

```python
# col wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=2)
```
