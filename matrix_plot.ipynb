#### Matrix Plot
- Heatmap
- Clustermap

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python

gap = px.data.gapminder()
gap
```

```python
# heatmap
# plot rectangular data as encoded matrrix 

temp_df=gap.pivot(index='country', columns='year', values='lifeExp')
temp_df
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
# axes level
```

```python
# annot 
temp_df=gap[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')
sns.heatmap(temp_df, annot=True, linewidths=1.5, cmap='summer')
```
