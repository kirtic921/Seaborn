#### Distribution plot
- used for univariate analysis
- used to find out the distribution
- range of the observation
- central tendency
- is the dat bimodal?
- are their outliers?

### HISTPLOT, KDEPLOT & RUGPLOT

```python
# fig level -> displot
# axes plot -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram

tips = sns.load_dataset('tips')
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
# bins parameter

sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

```python
# its also possible to visualize the distribution of categorical columns throu histograms using bins
# discrete bins are automatically set for categorical columns

sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue and element parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# facet plot using displot
sns.displot(data=tips, x='tip', kind='hist', col='sex', element = 'step')
```

```python
# kde Kernel density plot
# rather than using discrete bins, a kde plot smoothes the observations with a gaussian kernel, producinng a continous density estimate
sns.kdeplot(data=tips, x='total_bill')
```

```python
# hue, fill

sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True)
```
