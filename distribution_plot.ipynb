```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Distribution plot
- used for univariate analysis
- used to find out the distribution
- range of the observation
- central tendency
- is the dat bimodal?
- are their outliers?

### HISTPLOT, KDEPLOT & RUGPLOT

```python
# fig level -> displot
# axes plot -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram

tips = sns.load_dataset('tips')
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
# bins parameter

sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

```python
# its also possible to visualize the distribution of categorical columns throu histograms using bins
# discrete bins are automatically set for categorical columns

sns.displot(data=tips, x='day', kind='hist')
```
