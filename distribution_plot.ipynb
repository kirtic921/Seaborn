#### Distribution plot
- used for univariate analysis
- used to find out the distribution
- range of the observation
- central tendency
- is the dat bimodal?
- are their outliers?

### HISTPLOT, KDEPLOT & RUGPLOT

```python
# fig level -> displot
# axes plot -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram

tips = sns.load_dataset('tips')
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
# bins parameter

sns.displot(data=tips, x='total_bill', kind='hist', bins=20)
```

```python
# its also possible to visualize the distribution of categorical columns throu histograms using bins
# discrete bins are automatically set for categorical columns

sns.displot(data=tips, x='day', kind='hist')
```

```python
# hue and element parameter

sns.displot(data=tips, x='tip', kind='hist', hue='sex')
```

```python
sns.displot(data=tips, x='tip', kind='hist', hue='sex', element='step')
```

```python
# facet plot using displot
sns.displot(data=tips, x='tip', kind='hist', col='sex', element = 'step')
```

```python
# kde Kernel density plot
# rather than using discrete bins, a kde plot smoothes the observations with a gaussian kernel, producinng a continous density estimate
sns.kdeplot(data=tips, x='total_bill')
```

```python
# hue, fill
# aspect and height for changing the size of fig level plots
sns.displot(data=tips, x='total_bill', kind='kde', hue='sex', fill=True, aspect=2, height=10)
```

```python
# rugplot

# plot marginal distributions by drawing ticks along the x and y axes
# this function is intended to complement other plots by showing the location of indivisual observations in unobstructive way.
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
# bi variate histograms
# a bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```

```python
# bivariate kdeplots

# a bivariate kdeplot soothes the (x,y) observation with a 2D gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```
